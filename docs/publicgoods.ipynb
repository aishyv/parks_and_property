```python
#Scrape data from property site
url = "https://www.magicbricks.com/blog/property-rates-in-bangalore/130414.html"
rates_html = requests.get(url).text
```

```python
rates_soup= BeautifulSoup(rates_html)
```

```python
tables = rates_soup.select("table")
len (tables)
```

Of the 4 tables, there are 3 that are useful - Property rates in Affordable category, Mid segment and Premium range

```python
aff_table = tables[0]
aff_table.text[:500]
```

```python
mid_table = tables[1]
mid_table.text[:500]
```

```python
prem_table = tables[2]
prem_table.text[:500]
```

```python
aff_row = aff_table.select("tbody tr")
len(aff_row)
```

```python
mid_row = mid_table.select("tbody tr")
len(mid_row)
```

```python
prem_row = prem_table.select("tbody tr")
len(prem_row)
```

```python
aff_entries = [
    [ cell.text for cell in row.select("td") ]
    for row in aff_row
]

aff_entries[:3]
```

```python
mid_entries = [
    [ cell.text for cell in row.select("td") ]
    for row in mid_row
]

mid_entries[:3]
```

```python
prem_entries = [
    [ cell.text for cell in row.select("td") ]
    for row in prem_row
]

prem_entries[:3]
```

```python
df1 = pd.DataFrame(aff_entries, columns = aff_entries[2])
df1.head()
```

```python
df1 = df1.drop(0)
df1 = df1.drop(1)
df1 = df1.drop(2)
df1.head()
```

```python
df1 = df1.sort_values(by = ['Avg. Rate (INR per square feet)'])
df1.head()
```

```python
df2 = pd.DataFrame(mid_entries, columns = aff_entries[2])
df2 = df2.sort_values(by = ['Avg. Rate (INR per square feet)'])
df2.tail()
```

```python
df3 = pd.DataFrame(prem_entries, columns = prem_entries[2])
df3 = df3.drop(0)
df3 = df3.drop(1)
df3 = df3.drop(2)
df3.rename(columns = {'Avg. Rate (INR psf)':'Avg. Rate (INR per square feet)'},inplace = True)
df3 = df3.sort_values(by = ['Avg. Rate (INR per square feet)'])
df3.head()
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
df
```
